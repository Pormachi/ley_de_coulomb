```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#k: constante de Coulomb
k = 8.98e9

#Cargas:
q1 = 1e-6
q2 = -2e-6

#Distancia entre las cargas (en metros)
d = np.linspace(0.1, 10, 100)

#Fuerza Eléctrica (en Newtons)
F = k * q1 *q2 / (d ** 2)
```

```python
#Graficamos la Fuerza eléctrica vs la distancia
plt.plot(d, F)
plt.title('Ley de Coulomb')
plt.xlabel('Distancia (m)')
plt.ylabel('Fuerza Eléctrica (N)')
plt.show()
```
